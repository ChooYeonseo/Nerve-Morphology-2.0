# Manually Setting the Nerve Morphology

Due to high variations by nerve, it is best to set the morphology manually.

```python
import numpy as np
import math
import matplotlib.pyplot as plt
from matplotlib.patches import Polygon

from nmutilis.readcsv import helper_dict, raw_dict
from nmutilis.bone_model import Bone
from nmutilis.NerveNodule import Nerve, Node
```

```python
code_list = list(raw_dict.keys())
a = Bone(1, 2000)
code_list
```

```python
Nerve_list = []
```

# 23-21R Morphology

## Step1: Create Nerve Class

From the code list, assign a code. Then access the data

```python
code = code_list[0]
N23_21R = Nerve(code, raw_data=raw_dict[code], helper_data=helper_dict[code], bone=a)
```

## Step 2: Add relations manually

```python
N23_21R.add_relation('tip_cp', 'mid12_sn1')
N23_21R.add_relation('mid12_sn1', 'mid10_sn1')
N23_21R.add_relation('mid10_sn1', 'brch1')
N23_21R.add_relation('brch1', 'mid_sn1')
N23_21R.add_relation('brch1', 'mid_sn2')
N23_21R.add_relation('mid_sn1', 'tip_sn1')
N23_21R.add_relation('mid_sn2', 'tip_sn2')
```

```python
lines = N23_21R.get_branch_lines()
for line in lines:
    plt.plot(line[:,0], line[:,1])
    plt.plot(a.x, a.y)
plt.show()

Nerve_list.append(N23_21R)
```
